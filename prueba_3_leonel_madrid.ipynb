```python
def agregar_persona(lista_personas):
    rut = input("Ingrese su Rut: ")
    while True:
        nombre = input("Ingrese su nombre debe contener al menos 8 caracteres")
        if len(nombre) >= 8:
            break
        else:
            print("Nombre inválido ingrese un nombre con al menos 8 caracteres.")
    while True:
        fecha_nacimiento = input("Ingrese la Fecha de Nacimiento ejemplo (15/02/2004): ")
        break
    while True:
        edad = input("Ingrese su edad debe ser mayor a 0")
        if edad.isdigit() and int(edad) > 0:
            break
        else:
            print("Edad inválida ingrese un nuevamente su edad debe ser mayor a 0: ")
    while True:
        estado_civil = input("Ingrese si está soltero o casado: ")
        if estado_civil.lower() == "soltero" or estado_civil.lower() == "casado":
            break
        else:
            print("Estado Civil inválido. Por favor, ingrese soltero o casado ")
    persona = {"Rut": rut, "Nombre": nombre, "Fecha de Nacimiento": fecha_nacimiento, "Edad": edad, "Estado Civil": estado_civil}
    lista_personas.append(persona)
    print("sus datos se guardaron")


def imprimir_certificado_nacimiento(persona):
    print("CERTIFICADO DE NACIMIENTO")
    print("NIF:", persona["Rut"])
    print("Nombre:", persona["Nombre"])
    print("Fecha de Nacimiento:", persona["Fecha de Nacimiento"])
    print("Edad:", persona["Edad"])
    print("--------------------------------------")


def imprimir_certificado_estado_conyugal(persona):
    print("CERTIFICADO DE ESTADO CONYUGAL")
    print("NIF:", persona["Rut"])
    print("Nombre:", persona["Nombre"])
    print("Estado Civil:", persona["Estado Civil"])
    print("Edad:", persona["Edad"])
    print("-------------------------------------------")

def buscar_persona(lista_personas):
    rut_buscar = input("Ingrese el Rut para buscar sus datos  ")
    for persona in lista_personas:
        if persona["Rut"] == rut_buscar:
            print("Datos encontrados:")
            print("Rut:", persona["Rut"])
            print("Nombre:", persona["Nombre"])
            print("Edad:", persona["Edad"])
            return persona
    print("No se encontraron datos para el Rut ingresado.")
    return

def mostrar_menu():
    print("1. Agregar persona")
    print("2. Buscar persona")
    print("3. Imprimir certificado")
    print("4. Salir")

personas = []

while True:
    mostrar_menu()
    opcion = input("Ingrese el número de la opción deseada: ")

    if opcion == "1":
        agregar_persona(personas)
    elif opcion == "2":
        buscar_persona(personas)
    elif opcion == "3":
        persona_encontrada = buscar_persona(personas)
        if persona_encontrada:
            print("CERTIFICADOS DISPONIBLES ")
            print("1. Certificado de Nacimiento")
            print("2. Certificado de Estado Conyugal")
            certificado_opcion = input("Ingrese el número del certificado que desea imprimir: ")
            if certificado_opcion == "1":
                imprimir_certificado_nacimiento(persona_encontrada)
            elif certificado_opcion == "2":
                imprimir_certificado_estado_conyugal(persona_encontrada)
            else:
                print("Opción inválida. No se pudo imprimir el certificado.")
    elif opcion == "4":
        print("adios")
        break
```
